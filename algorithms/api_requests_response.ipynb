### Pagination

helps you get different sets of records at ones. where there is a limit of api calls that you can make

```python
url = "https://api.fda.gov/food/enforcement.json"
dfs = [] #collects the dataframes

#iterating over the range to fetch 25000 records in batches of 1000
for i in range(25):
    params = {
        "limit": 1000,
        "skip" : i * 1000
    }

    #fetching data from API
    data = requests.get(url,params = params).json()
    dfs.append(pd.DataFrame(data['results']))#append dataframe of 1000 rows to list of dfs
#combine dataframes
df = pd.concat(dfs, ignore_index = True) #ignore index otherwise it will combine to same rows
```

### Normalization

```python
#calculate each value minus min / range
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
normalize(df['days_since_last_inspection'])
```

```python
#lower value the better
df['days_normalized'] = 1-normalize(df['days_since_last_inspection'])
```

```python
df['violations_normalized'] = 1 - normalize(df['scaled_violations'])
```

```python
df['violations_normalized'].hist()
```

```python
df['scaled_violations'].hist()
```

```python
df["composite_score"] = (0.3 * df['days_normalized']
                          + 0.7 *  df['violations_normalized']) * 100
```

```python
df['composite_score'] = df['composite_score'].round(2)
```
